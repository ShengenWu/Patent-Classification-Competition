```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Academic-paper-classification/dataset/testA_zh_en_final.csv')
df.head()
```

```python
train_text_list = []
for index, row in df.iterrows():
    abstract = row["abstract_en"]
    split = abstract.split('.')
    statement = split[0]
    effect = split[-2]
    TEM = statement + effect
    programme_list = [word for word in split if word not in TEM]
    programme = ' '.join(programme_list)
    statement = statement + '.'
    effect = effect + '.'
    programme = programme + '.'
    text = statement + " [SEP] " + programme + " [SEP] " + effect
    train_text_list.append(text)
df['abstract_en'] = train_text_list
```
